```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
sns.set()
plt.rcParams['figure.figsize'] = [16, 10]
```

```python
df = pd.read_csv("recall_per_sample_pair.tsv", sep="\t")
df
```

```python
df["label"] = df.apply(lambda row: f"Probes from {row['sample']}\nComparison {row['sample_pair']}\nAligned bases {row['aligned_bases_percentage']}%", axis=1)
df
```

```python
# order based on aligned_bases_percentage
ordered_df = df.sort_values(by="aligned_bases_percentage", ascending=False)
ordered_df
```

```python
# order based on snippy recall
label_order = df[df.tool=="snippy_NZ_CP016497.1"].sort_values(by="recall", ascending=False)["label"].tolist()
ordered_df = df
ordered_df["order"] = ordered_df["label"].apply(
    lambda label: label_order.index(label)
)
ordered_df.sort_values(by="order", inplace=True)
ordered_df
```

```python
lineplot = sns.lineplot(x="label", y="recall",
             hue="tool", data=ordered_df, sort=False)
plt.xticks(rotation=90)
lineplot
```

```python
lineplot = sns.lineplot(x="label", y="recall",
             hue="tool",
            data=ordered_df[ordered_df.apply(
                lambda row: "063_STEC" not in row["sample_pair"], axis=1)],
            sort=False)
plt.xticks(rotation=90)
lineplot
```

```python
lineplot = sns.lineplot(x="label", y="recall",
             hue="tool",
            data=ordered_df[ordered_df.apply(
                lambda row: "063_STEC" in row["sample_pair"], axis=1)],
            sort=False)
plt.xticks(rotation=90)
lineplot
```
