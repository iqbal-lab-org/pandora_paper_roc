```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
# import jtplot submodule from jupyterthemes
from jupyterthemes import jtplot

# currently installed theme will be used to
# set plot style if no arguments provided
jtplot.style()
```

```python
def parse_raw_svlen(raw_stats, sample_name):
    """Parses raw stats from SVLEN file that has just been copied and pasted into a string.
    Args:
        raw_stats (str): A string representation of the data copied and pasted as is from 
        the *.info files generated from vcf_stats.sh script.
        sample_name (str): Name of the sample the stats are for.
        
    Returns:
        (pd.DataFrame): A Pandas DataFrame object with columns for the SV length, counts 
        for the length , and a column of the sample name.
    """
    counts = []
    lengths = []
    for x in raw_stats.strip().split('\n'):
        count, length = x.split()
        counts.append(int(count))
        lengths.append(int(length))
        
    return pd.DataFrame({
        "counts": counts,
        "lengths": lengths,
        "sample": [sample_name] * len(counts)
    })
    
```

```python
def plot_svlen(df, figsize=(18, 12), x_log=False, y_log=False):
    """Generate a barplot of SV length count data.
    Args:
        df (pd.DataFrame): DataFrame with counts in column named 'counts'
        and lengths in column named 'lengths'.
        figsize (tuple[int, int]): Figure size of (width, height) in inches.
        x_log (bool): Plot x-axis on log-scale?
        y_log (bool): Plot y-axis on log-scale?
    """
    plt.figure(figsize=figsize)
    p = sns.barplot(x='lengths', y='counts', data=df, color='b')
    if x_log:
        p.set_xscale('log')
    if y_log:
        p.set_yscale('log')
    sns.despine()
    return p
```
