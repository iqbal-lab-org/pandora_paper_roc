```python
import pysam
import os
```

```python
test_bam_path = os.path.abspath('../../nanotest/alignment.bam')
```

```python
with pysam.AlignmentFile(test_bam_path, 'rb') as samfile:
    print(samfile.references)
    for pileupcolumn in samfile.pileup('NC_016845.1'):
        if pileupcolumn.reference_pos < 100:
            continue
        elif pileupcolumn.reference_pos > 120:
            break

            
        print ("\ncoverage at base %s = %s" %
           (pileupcolumn.pos, pileupcolumn.n))
        for pileupread in pileupcolumn.pileups:
            if not pileupread.is_del and not pileupread.is_refskip:
                # query position is None if is_del or is_refskip is set.
                print ('\tbase in read %s = %s' %
                      (pileupread.alignment.query_name,
                       pileupread.alignment.query_sequence[pileupread.query_position]))
```

```python
with pysam.AlignmentFile(test_bam_path, 'rb') as samfile:
    i = 0
    for read in samfile:
        if i > 10: break
        positions = read.get_aligned_pairs()
        print(positions)
#         print("Start: {}\tEnd: {}".formamap(min(positions), max(positions)))
        i += 1
```
