## Plotting mapping quality 
This is the mapping quality from mapping the truth probes to the vcf probes for a query sample. What we are looking for is some kind of indication as to whether there is a clear "inflection" point where we can filter mappings to remove spurrious alignments/multi-mappings.

```python
import seaborn as sns
from pathlib import Path
import pysam
import matplotlib.pyplot as plt
FIGSIZE = (13, 9)
```

```python
samfile = Path("../tmp/CFT073.panel_to_probes.sam")
```

```python
mapping_quality_unmapped = []
mapping_quality_secondary = []
mapping_quality_primary = []
with pysam.AlignmentFile(samfile) as sam:
    for record in sam:
        score = record.mapping_quality
        if not record.is_unmapped:
            if not record.is_secondary and not record.is_supplementary:
                mapping_quality_primary.append(score)
            else:
                mapping_quality_secondary.append(score)
        else:
            mapping_quality_unmapped.append(score)
```

```python
# query flank = 50
# truth flank = 25
xs = [(mapping_quality_unmapped, "Unmapped"), (mapping_quality_secondary, "Secondary"), (mapping_quality_primary, "Primary")]
qualities, labels = zip(*xs)
fig, ax = plt.subplots(figsize=FIGSIZE)
ax.hist(qualities, label=labels)
ax.legend()
# for x, label in xs:
#     sns.distplot(x, ax=ax, label=label)
ax.set(title="Mapping quality for truth probes mapped to VCF probes", xlabel="Mapping quality")
```
