```python
import pandas as pd
import numpy as np
```

```python
df_pvr = pd.read_csv("recall_per_nb_of_samples_pvr.plot_data.csv")
df_pvr = df_pvr[["tool", "NB_OF_SAMPLES", "recall_PVR"]]
df_pvr
```

```python
df_avgar = pd.read_csv("recall_per_nb_of_samples_avgar.plot_data.csv")
df_avgar = df_avgar[["tool", "NB_OF_SAMPLES", "recall_AvgAR"]]
df_avgar
```

```python
merged_df = df_pvr.merge(df_avgar, on=["tool", "NB_OF_SAMPLES"])
merged_df
```

```python
merged_df["difference"] = np.abs(merged_df["recall_AvgAR"]-merged_df["recall_PVR"])
merged_df.to_csv("debugging_recall_PVR_vs_AvgAR.csv")
merged_df
```
