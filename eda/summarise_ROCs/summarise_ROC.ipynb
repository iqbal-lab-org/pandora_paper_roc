```python
import pandas as pd
import numpy as np
```

```python
technologies=["illumina", "nanopore"]
filters=["", "_with_filters"]

def remove_ref_from_tool_name(tool_name):
    tools = ["snippy", "samtools", "nanopolish", "medaka"]
    for tool in tools:
        if tool_name.startswith(tool):
            return tool
    return tool_name

for tech in technologies:
    for filter in filters:
        df = pd.read_csv(f"pandora1_paper_analysis_output_20_way/{tech}_analysis{filter}/ROC_data.tsv", sep="\t")
        df = df[df.step_GT==0]
        df = df.rename(columns={"recalls_wrt_truth_probes": "TAR",
                                "recalls_wrt_variants_where_all_allele_seqs_were_found": "PVR",
                                "recalls_wrt_variants_found_wrt_alleles": "AvgAR"})
        df = df[["tool", "coverage", "coverage_threshold", "strand_bias_threshold", "gaps_threshold", "error_rate", "TAR", "PVR", "AvgAR"]]
        
        df["tool"] = df["tool"].apply(remove_ref_from_tool_name)
        
        summarised_df = df.groupby(["tool", "coverage", "coverage_threshold", "strand_bias_threshold", "gaps_threshold"]).agg({
            "error_rate": [np.min, np.max],
            "TAR":        [np.min, np.max],
            "PVR":        [np.min, np.max],
            "AvgAR":      [np.min, np.max],
        })
        summarised_df.columns = ["_".join(x) for x in summarised_df.columns.ravel()]
        display(summarised_df)
        summarised_df.to_csv(f"{tech}_analysis{filter}.summarised_ROC_data.csv")
```
