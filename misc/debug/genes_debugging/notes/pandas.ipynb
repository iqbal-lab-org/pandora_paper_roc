```python
import pandas as pd
```

```python
df = pd.read_csv("/home/leandro/git/pandora1_paper/notes/063_STEC_and_CFT073.report.tsv", sep="\t")
all_truth_probes = df.query_probe_header.unique()
all_truth_probes
```

```python
def get_best_mapping_for_truth_probe(truth_probe):
    all_mappings_for_the_truth_probe = df.query("query_probe_header == @truth_probe")
    correct_classification_query_str = "classification == 'primary_correct' or classification == 'secondary_correct' or classification == 'supplementary_correct'"
    mappings_with_correct_classifications = all_mappings_for_the_truth_probe.query(correct_classification_query_str)
    if len(mappings_with_correct_classifications) > 0:
        # just select any correct mapping, which is a TP
        return mappings_with_correct_classifications.iloc[0]
    else:
        # just select any mapping, which are all FP
        return all_mappings_for_the_truth_probe.iloc[0]

def get_best_mapping_all_truth_probes(all_truth_probes):
    truth_probe_to_best_mapping = {}
    for truth_probe in all_truth_probes:
        truth_probe_to_best_mapping[truth_probe] = get_best_mapping_for_truth_probe(truth_probe)
    return truth_probe_to_best_mapping

def get_df_with_best_mapping_for_all_truth_probes():
    truth_probe_to_best_mapping = get_best_mapping_all_truth_probes(all_truth_probes)
    return pd.DataFrame(data=truth_probe_to_best_mapping.values())

new_df = get_df_with_best_mapping_for_all_truth_probes()
display(new_df)
```
